```python
import json

def prompt_v1(review_text):
    prompt = f"""
    You are an assistant that classifies Yelp reviews.

    Read the review and predict a star rating from 1 to 5.

    Return ONLY valid JSON in this exact format:

    {{
      "predicted_stars": <number from 1 to 5>,
      "explanation": "brief reason for the rating"
    }}

    Review:
    {review_text}
    """

    response = model.generate_content(prompt)

    try:
        return json.loads(response.text)
    except:
        return {"predicted_stars": None, "explanation": "Invalid JSON"}
```
